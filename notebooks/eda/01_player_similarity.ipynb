# Player Similarity — Content-Based Recommendation

Build a FAISS index from player bio embeddings, then find the most
similar players for a given player. Uses `src.models.similarity`.


```python
target_player = "alcaraz"
top_k = 5
```

```python
from src.models.similarity import (
    build_index,
    PlayerSimilarity,
    DEFAULT_INDEX,
    DEFAULT_NAMES,
    DEFAULT_DISPLAY_NAMES,
)

build_index()
```

```python
finder = PlayerSimilarity()

if target_player not in finder:
    resolved = finder.find_by_name(target_player)
    if resolved:
        target_player = resolved
        print(f"Resolved to player_id: {target_player}")
    else:
        print(f"Player '{target_player}' not found in index.")
        print(f"Available: {', '.join(finder.player_ids)}")

similar = finder.similar(target_player, top_k=top_k)
print(f"\nPlayers similar to {target_player}:")
for s in similar:
    print(f"  {s['player_id']:12s}  {s['display_name']}")
```
