# Player Similarity — Content-Based Recommendation

Build a FAISS index from player bio embeddings, then find the most
similar players for a given player. Uses `src.models.similarity`.


```python
target_player = "carlos alcaraz"
top_k = 5
```

```python
from src.models.similarity import PlayerSimilarity

similarity_index = PlayerSimilarity()
similarity_index.build()
```

```python
similar_players = similarity_index.search(target_player, top_k=top_k)

if not similar_players:
    print(f"Player '{target_player}' not found.")
else:
    print(f"\nPlayers similar to {target_player}:")
    for s in similar_players:
        print(f"  {s['player_id']:12s}  {s['display_name']:25s}  {s['score']}")
```
